```python
import matplotlib.pyplot as plt
from matplotlib import style
import random
import numpy as np
import math
import scipy.signal
```

```python
def anel(vertices, m):
    links = np.empty((1, 2)) 
    for i in range(0, vertices, 1):
        for n in range(-m, m + 1, 1):
            if i+n < i:
                # print(i, i+n)
                links = np.append(links, [[int(i), int(i+n)]], axis=0)
    return np.delete(links, 0, axis=0)

def rewire(links, p):
    for i in range(0, len(links), 1):
        if links[i,1] < 0:
            links[i,1]=vertices+links[i,1]
        elif links[i,0] < 0:
            links[i,0]=vertices+links[i,0]
        # print(i)
        f = 0
        r = np.random.random()
        if r < p:
            while f < 1:
                nodonovo = np.random.randint(0, high=vertices)
                if np.any(links[:, 1] == nodonovo) and np.any(links[:, 0] == links[i, 1]):
                    # print(links[j, :])
                    # print(links[i, :])
                    # print('-------------')
                    f = 0
                    break
                elif np.any(links[:, 0] == nodonovo) and np.any(links[:, 1] == links[i, 0]):
                    # print(links[j, :])
                    # print(links[i, :])
                    # print('-------------')
                    f = 0
                    break
                elif np.any(links[:, 1] == links[i, 1]) and np.any(links[:, 0] == nodonovo):
                    f = 0
                    break
                elif np.any(links[:, 0] == links[i, 0]) and np.any(links[:, 1] == nodonovo):
                    f = 0
                    break
                else:
                    f = 1
                if links[i, 1] == nodonovo:
                    f = 0
                else:
                    f = f
            links[i, 0] = nodonovo
    return links

def cluster(vertices, links):
    c = np.empty(1)
    for i in range(0, vertices, 1):
        # print(i)
        count = 0
        vizinhos = np.empty(1)
        for j in range(0, len(links), 1):
            if links[j,0] == i:
                vizinhos = np.append(vizinhos, links[j,1])
            elif links[j,1] == i:
                vizinhos = np.append(vizinhos, links[j,0])
        vizinhos = np.delete(vizinhos, 0)
        # print(vizinhos)
        # mudar isso aqui com np.any
        for j in range(0, len(links), 1):
            for k in range(0, len(vizinhos), 1):
                for m in range(0, len(vizinhos), 1):
                    if vizinhos[k] == links[j, 0] and vizinhos[m] == links[j, 1]:
                        # print(links[j, :])
                        # print(vizinhos[k], vizinhos[m])
                        count +=1
        # print(count)
        count = 2 * count / (len(vizinhos) * (len(vizinhos)-1))
        c = np.append(c, count)
    return np.average(np.delete(c, 0))
```

```python
INF = 999

def matriz(vertices, links):
    G = np.zeros((vertices, vertices))
    for i in range(len(links)):
        G[int(links[i, 0]), int(links[i, 1])] = 1
        G[int(links[i, 1]), int(links[i, 0])] = 1
    for i in range(vertices):
        for j in range(vertices):
            if G[i, j] < 1:
                G[i, j] = INF
    return G
       
def floyd(G):
    dist = list(map(lambda p: list(map(lambda q: q, p)), G))
    for r in range(vertices):
        for p in range(vertices):
            for q in range(vertices):
                dist[p][q] = min(dist[p][q], dist[p][r] + dist[r][q])
    # sol(dist)
    # print(dist)
    dist = np.array(dist)
    # return(dist)
    return np.sum(dist)/(vertices * (vertices-1))


def sol(dist):    
    for p in range(vertices):
        for q in range(vertices):
            if(dist[p][q] == INF):
                print("INF", end=" ")
            else:
                print(dist[p][q], end="  ")
        print(" ")
```
